# Training from scratch

```python
import os
from nltk import sent_tokenize
import nltk
nltk.download('punkt')
from gensim.utils import simple_preprocess
#Download below dataset from this link "https://www.kaggle.com/datasets/khulasasndh/game-of-thrones-books/data"
story = []
for filename in os.listdir('data'):
    f = open(os.path.join('data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
story
```

```python
# Training the Word2Vec model
#w2v_model = gensim.models.Word2Vec(story,window=10,min_count=2)
#w2v_model=gensim.models.Word2Vec(window=10,min_count=2)
model = Word2Vec(story, window=10,min_count=2)
```

```python
# Getting vocabulary of wards.
#model.wv.vocab
model.wv.index_to_key
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```
