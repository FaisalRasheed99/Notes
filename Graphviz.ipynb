```python
from graphviz import Digraph

dot = Digraph()

# Defining nodes
dot.node('Start', 'Start')
dot.node('Upload', 'User Uploads PDF and Enters Query')
dot.node('Process', 'Process File and Query')
dot.node('Check', 'PDF and Query Provided?')
dot.node('Error', 'Error Message: "Please upload a PDF and enter a query."')
dot.node('Load', 'Load PDF and Create Retriever')
dot.node('Query', 'Process Query')
dot.node('Similarity', 'Similarity Search')
dot.node('RetrievalQA', 'RetrievalQA')
dot.node('Chain', 'Conversation Chain')
dot.node('Update', 'Update and Format Conversation History')
dot.node('Display', 'Display Answer and Conversation History')
dot.node('Interact', 'User Interacts Again or Ends Session')
dot.node('Submit', 'Submit Another Query')
dot.node('Refresh', 'Refresh and Clear History')
dot.node('End', 'End')

# Defining edges
dot.edge('Start', 'Upload')
dot.edge('Upload', 'Process')
dot.edge('Process', 'Check')
dot.edge('Check', 'Error', label='No')
dot.edge('Check', 'Load', label='Yes')
dot.edge('Load', 'Query')
dot.edge('Query', 'Similarity')
dot.edge('Query', 'RetrievalQA')
dot.edge('Query', 'Chain')
dot.edge('Similarity', 'Update')
dot.edge('RetrievalQA', 'Update')
dot.edge('Chain', 'Update')
dot.edge('Update', 'Display')
dot.edge('Display', 'Interact')
dot.edge('Interact', 'Submit', label='Submit Another Query')
dot.edge('Interact', 'Refresh', label='Refresh')
dot.edge('Submit', 'Start')
dot.edge('Refresh', 'Start')
dot.edge('Interact', 'End', label='End')

# Render the diagram
dot.render('flow_diagram', format='png', cleanup=True)
```
