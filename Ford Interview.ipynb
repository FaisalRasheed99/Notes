```python
l = [4,9,2,6,1]
```

```python
def add(l):
    if len(l) == 0:
        return 0
    else :
        return l[0] + add(l[1:])
```

```python
add(l)
```

```python
class Audi():
    def col(self):
        print("color is blue.")
        
class BMW():
    def col(self):
        print("color is red.")
        
class Ferrari():
    def col(self):
        print("color is green.")
```

```python
def prop(obj):
    obj.col()
    
```

```python
obj_Audi = Audi()
obj_BMW = BMW()
obj_Ferrari = Ferrari()
```

```python
prop(obj_Audi)
```

```python
prop(obj_BMW)
```

```python
prop(obj_Ferrari)
```
