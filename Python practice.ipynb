```python
# Recursive function to check if a string is palindrome
def isPalindrome(s):
    l = len(s)
    if l < 2:
        return True
    elif s[0] == s[l - 1]:
        return isPalindrome(s[1: l - 1])
    else:
        return False
s = "MalaYaLam"
ans = isPalindrome(s)
if ans:
    print("Yes")
else:
    print("No")
```

```python
#Write a program that accepts 2 numbers from the user a numerator and a denominator and then simplifies it Eg if the num = 6, den = 9 the answer should be ⅔
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def simplify_fraction(numerator, denominator):
    common_factor = gcd(numerator, denominator)
    simplified_numerator = numerator // common_factor
    simplified_denominator = denominator // common_factor
    return simplified_numerator, simplified_denominator

try:
    # Accept numerator and denominator from the user
    numerator = int(input("Enter the numerator: "))
    denominator = int(input("Enter the denominator: "))

    # Simplify the fraction
    simplified_numerator, simplified_denominator = simplify_fraction(numerator, denominator)

    # Print the simplified fraction
    print(f"The simplified fraction is: {simplified_numerator}/{simplified_denominator}")
except ValueError:
    print("Please enter valid integers for the numerator and denominator.")
except ZeroDivisionError:
    print("Denominator cannot be zero. Please enter a non-zero denominator.")
```

```python
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?"
]
def preprocess_text(text):
    cleaned_text = ''.join(char.lower() if char.isalpha() or char.isspace() else ' ' for char in text)
    return cleaned_text
def create_bow(texts):
    tokenized_texts = [preprocess_text(text).split() for text in texts]
    vocabulary = set(word for text in tokenized_texts for word in text)
    bow_vectors = [{word: text.count(word) for word in vocabulary} for text in tokenized_texts]
    return bow_vectors, list(vocabulary)
numerical_vectors, feature_names = create_bow(documents)
print("Numerical Vectors:")
for vector in numerical_vectors:
    print(list(vector.values()))
print("\nFeature Names:")
print(feature_names)
```

```python
def lcm(a, b):
    max_num = max(a, b)
    min_num = min(a, b)
    multiple = max_num
    while True:
        if multiple % min_num == 0:
            return multiple
        multiple += max_num
num1 = 12
num2 = 18
result = lcm(num1, num2)
print(f"The LCM of {num1} and {num2} is: {result}")
```

```python
#matrix multiplication
matrix_a = [[1, 2, 3], [4, 5, 6]]
matrix_b = [[7, 8], [9, 10], [11, 12]]

rows_a, cols_a = len(matrix_a), len(matrix_a[0])
rows_b, cols_b = len(matrix_b), len(matrix_b[0])

# Initialize the result matrix with zeros
result = [[0 for _ in range(cols_b)] for _ in range(rows_a)]

for i in range(rows_a):
    for j in range(cols_b):
        for k in range(cols_a):
            result[i][j] += matrix_a[i][k] * matrix_b[k][j]

result
```

```python
def find_max(nested_list):
    max_value = nested_list[0][0]
    for sublist in nested_list:
        for element in sublist:
            if element > max_value:
                max_value = element
    return max_value


# Example usage:
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
max_value = find_max(nested_list)
print(max_value)
```

```python
#nested list
def flatten(list1):
    flat_list = []
    for element in list1:
        if type(element) is list:
            for item in flatten(element):
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list

list1 = [[1, 2, 3], [4, 5], [6, 7, 8]]
print(flatten(list1))
```

```python
#find minimum number of rotations to obtain actual string
def findRotations(str1, str2):
    # To count left rotations of string
    x = 0
    # To count right rotations of string
    y = 0
    m = str1
    while True:
        # left rotating the string
        m = m[len(m)-1] + m[:len(m)-1] 
        # checking if rotated and actual string are equal.
        if(m == str2):
            x += 1
            break
        else:
            x += 1
            if x > len(str2) :
                break
    while True:
        # right rotating the string
        str1 = str1[1:len(str1)]+str1[0] 
        # checking if rotated and actual string are equal.
        if(str1 == str2):
            y += 1
            break
        else:
            y += 1
            if y > len(str2):
                break
    if x < len(str2):
        # printing the minimum number of rotations.
        print(min(x,y))
    else:
        print("given strings are not of same kind")
findRotations('sgeek', 'geeks')
```

```python
def isPrime(n):
    if(n <= 1):
        return False
    if(n <= 3):
        return True
    # This is checked so that we can skip middle five numbers in below loop 
    if(n % 2 == 0 or n % 3 == 0):
        return False
    for i in range(5,int((n**0.5) + 1), 6): 
        if(n % i == 0 or n % (i + 2) == 0):
            return False
    return True

def nextPrime(N):
    if (N <= 1):
        return 2
    prime = N
    found = False
    while(not found):
        prime = prime + 1
        if(isPrime(prime) == True):
            found = True
    return prime

nextPrime(10)
```

```python
#Define a class named Shape and its subclass Square. The Square class has an init function which takes a length as argument. Both classes have an area function whichcan print the area of the shape where Shape's area is 0 by default.
class Shape:
    def __init__(self):
        pass
    # Default constructor, no need to initialize anything
    def area(self):
        return 0
# Shape's area is 0 by default
class Square(Shape):
    def __init__(self,length):
        super().__init__()
        # Call the constructor of the parent class
        self.length=length
    def area(self):
        return self.length**2# Calculate the area of the square

# Create instances of the classes
shape=Shape()
square=Square(float(input("Enter the shape of the square: ")))
# Calculate and print the areas
print("Shape's area by default:",shape.area())
print("Area of the square:",square.area())
    
```

# non_local

To modify the local variable inside the local scope, we use global keyword, therefore, to modify the enclosed variable in the local scope, we use nonlocal keyword.

```python
y=10
def outer():
    z=15
    def inner():
        x=4
        nonlocal z
        z+=1
        print("x:",x)
        print("inside the function z : ",z)
    inner()
    print("z:",z)
outer() 
```

# LEGB Rule

 Python will first search in the local scope. If it is not found, then it will search in the enclosed group. If it is not found, it will search in the global scope. If it is not found, it will search in the built-in scope. Then also it is not found, it will give name error.

```python
#Python will first search in the local scope
x=5
def function():
    x=10
    def inner():
        x=15
        print("x:",x)
    inner()
function()
```

```python
#If it is not found, then it will search in the enclosed group
x=5
def function():
    x=10
    def inner():
        #x=15
        print("x:",x)
    inner()
function()
```

```python
#If it is not found, it will search in the global scope
x=5
def function():
    #x=10
    def inner():
        #x=15
        print("x:",x)
    inner()
function()
```

```python
def multiplier_outer(n):
    def multiplier_inner(x):
        return x * n
    return multiplier_inner

times3 = multiplier_outer(3)
times5 = multiplier_outer(5)
print(times3(9))
print(times5(3))
print(times5(times3(2)))
```

```python
import threading
import time
def f1(n):
    for i in range(n):
        print("Faisal")
        time.sleep(0.0001)
def f2(m):
    for i in range(m):
        print("Rasheed")
        time.sleep(0.0001)
if __name__ =="__main__":
    t1 = threading.Thread(target=f1, args=(10,))
    t2 = threading.Thread(target=f2, args=(10,))
    t1.start()
    t2.start()
    t1.join()
    t2.join()
```

# Deep Copy & Shallow Copy

#### For nested list

```python
# =
lst1=[[1,2,3],[3,4,5]]
lst2=lst1
lst2[1][0]=100
print(id(lst1),id(lst2))
lst2,lst1
```

```python
#copy
lst1=[[1,2,3],[3,4,5]]
lst2=lst1.copy()
lst2[1][0]=100
print(id(lst1),id(lst2))
lst2,lst1
```

```python
#deepcopy()
import copy
lst1=[[1,2,3],[3,4,5]]
lst2=copy.deepcopy(lst1)
lst2[1][0]=100
print(id(lst1),id(lst2))
lst2,lst1
```

#### For list ----> In a normal list shallow =deep copy

```python
## =
lst1=[1,2,3,4]
lst2=lst1
lst2[1]=1000
print(id(lst1),id(lst2))
lst2,lst1
```

```python
## Shallow Copy
lst1=[1,2,3,4]
lst2=lst1.copy()
lst2[1]=1000
print(id(lst1),id(lst2))
lst2,lst1
```

```python
##deep copy
import copy
lst1=[1,2,3,4]
lst2=copy.deepcopy(lst1)
lst2[1]=100
print(id(lst1),id(lst2))
lst2,lst1
```

##### If we append any new list inside the nested list, then it wo'nt be reflected in the original list. Whether if we do Shallow copy or deep copy.Will be reflected only in '='

```python
lst1=[[1,2,3,4],[5,6,7,8]]
lst2=lst1
lst1.append([9,10,11,12])
print(id(lst1),id(lst2))
lst1,lst2
```

```python
## Shallow Copy
lst1=[[1,2,3,4],[5,6,7,8]]
lst2=lst1.copy()
lst1.append([9,10,11,12])
print(id(lst1),id(lst2))
lst1,lst2
```

```python
##deep copy
lst1=[[1,2,3,4],[5,6,7,8]]
lst2=copy.deepcopy(lst1)
lst1.append([9,10,11,12])
print(id(lst1),id(lst2))
lst1,lst2
```
