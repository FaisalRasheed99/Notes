```python
#sBecause of Super, there is no need of defining. variables, methods or properties. inside the child class. constructor which has been already defined in the parent class. We can directly inherit them using this super function.
class OTTSubscription:
    def __init__(self,subscription_id,plan,payment):
        self.id = subscription_id
        self.plan = plan
        self.payment = payment
    def subscribe(self):
        print(f"Subscriber with {self.id} id subscribed to the {self.plan} plan")
    def unsubscribe(self):
        print(f"Subscriber with {self.id} id unsubscribed to the {self.plan} plan")           
netflix = OTTSubscription(121,"monthly",100)
print(netflix.plan)

print(netflix.subscribe())

class PremiumSubscription(OTTSubscription):
    def __init__(self,subscription_id,plan,payment,screens):
        self.id = subscription_id
        self.plan = plan
        self.payment = payment
        self.max_screens = screens
    def subscribe(self):
        print(f"Subscriber with {self.id} id subscribed to the {self.plan} plan")
    def unsubscribe(self):
        print(f"Subscriber with {self.id} id unsubscribed to the {self.plan} plan")           
    def set_max_screens(self,screens):
        self.max_screens = screens
        print(f"maximum screen set to {self.max_screens} in the Premium Plan")
netflix = PremiumSubscription("123","monthly",1,1200)
print(netflix.subscribe())

print(netflix.set_max_screens(4))

#final code
class PremiumSubscription(OTTSubscription):
    def __init__(self,subscription_id,plan,payment,screens):
        super().__init__(subscription_id,plan,payment)
        self.max_screens = screens
    def set_max_screens(self,screens):
        self.max_screens = screens
        print(f"maximum screen set to {self.max_screens} in the Premium Plan")
```

```python
# program to illustrate protected access modifier in a class
class Student:
    def __init__(self, name, roll): 
        self._name = name
        self._roll = roll
    def _displayRoll(self):
        print("Roll: ", self._roll)
class Geek(Student):
    def __init__(self, name, roll): 
        Student.__init__(self, name, roll) 
    def displayDetails(self):
        # accessing protected data members of super class 
        print("Name: ", self._name) 
        # accessing protected member methods of super class 
        self._displayRoll()
obj = Geek("Faisal", 1706256) 
obj.displayDetails()
```

```python
#protected variables of the class are also accessible outside the class environment.
#The protected variables and functions can be accessed by the derived classes and any other classes but it is the sole responsibility of the good programmer not to use the protected data members outside the specific class environment.
#Python does not have strict enforcement of private members, so they can still be accessed from outside the class.
class Anime:
    _rating = "Best"
    def __init__(self, anime, release_year):
        self._anime = anime
        self._release_year = release_year
    def _show(self):
        print("Release:", self._release_year)
obj = Anime("Naruto", 2002)
print("Anime:", obj._anime)
obj._show()
```

```python
class Super:
     def __init__(self, var1, var2, var3): 
        self.var1 = var1
        self._var2 = var2
        self.__var3 = var3
    # public member method
     def displayPublicMembers(self):
        # accessing public data members
        print("Public Data Member: ", self.var1)
    # protected member method
     def _displayProtectedMembers(self):
        # accessing protected data members
        print("Protected Data Member: ", self._var2)
    # private member method
     def __displayPrivateMembers(self):
        # accessing private data members
        print("Private Data Member: ", self.__var3)
    # public member method
     def accessPrivateMembers(self):	 
        # accessing private member method
        self.__displayPrivateMembers()
class Sub(Super):
    def __init__(self, var1, var2, var3): 
        Super.__init__(self, var1, var2, var3)
    # public member function 
    def accessProtectedMembers(self):
        # accessing protected member functions of super class 
        self._displayProtectedMembers()
obj = Sub("Geeks", 4, "Geeks !") 
obj.displayPublicMembers()
obj.accessProtectedMembers()
obj.accessPrivateMembers() 
print("Object is accessing protected member:", obj._var2)
# object can not access private member, so it will generate Attribute error
obj.__var3
```

```python
#Will give error because last method will be considered and last method takes three arguments and here only two arguments are passed.
a.add(10,20)
```

```python
a.add(10,20,30)
```

To prevent method overloading. We will join the two methods which were declared for two different purposes in such a way that the only single method will handle the function of both the method so that the purpose of 1st function will be utilised.

```python
class Addition:
    def add(self, num1=None, num2=None, num3=None) :
        if num1 != None and num2 != None and num3 != None:
            print("Addition is:",num1+num2+num3)
        elif num1 != None and num2 != None:
            print("Addition is:",num1+num2)
        else:
            print("incorrect parameters provided")

c1=Addition()
c1.add(10,20)
c1. add(10,20,30)
```

# @Property Decorator

By using @property, we can treat a method as An attribute

```python
#The derived attribute remains unchanged even if we change any attribute inside the constructor.
class Student:
    def __init__(self,name, grade) :
        self.name = name
        self.grade = grade
        self.msg = self.name + " got grade " + self.grade

stud1 = Student("nia","B")
stud1.name = "tom"    # Here we can see that if. we have changed the attribute inside the constructor that derived attribute remains unchanged.
print("name:",stud1.name)   
print("grade:",stud1.grade)
print(stud1.msg)
```

```python
class Student:
    def __init__(self,name, grade) :
        self.name = name
        self.grade = grade
    @property    #getter
    def msg (self) : #Here we can see by defining that derived attribute inside a method and declaring @property above that method it will behave as an attribute, not a method
        return self.name + " got grade " + self.grade

stud1 = Student("nia","B")
stud1.name = "tom"
print("name:",stud1.name)
print("grade:",stud1.grade)
print(stud1.msg)
```

```python
class Student:
    def __init__(self,name, grade) :
        self.name = name
        self.grade = grade
    @property  #getter
    def msg (self) :
        return self.name + " got grade " + self.grade
    @msg.setter   #setter
    def msg(self,msg) :
        sent = msg.split(" ")
        print(sent)
        self.name = sent[0]
        self.grade = sent[-1]

stud1 = Student("nia","B")
stud1.msg = "amulya got grade A"
print("name:",stud1.name)
print("grade:",stud1.grade)
print(stud1.msg)
```

```python
from abc import ABC, abstractmethod
class Interface(ABC):
    @abstractmethod
    def get_title(self):
        pass
    @abstractmethod
    def get_label_text(self):
        pass
    @abstractmethod
    def on_button_click(self):
        pass
class SimpleInterface(Interface):
    def get_title(self):
        return "Simple Interface"
    def get_label_text(self):
        return "Enter some text:"
    def on_button_click(self):
        user_input = input("Enter some text: ")
        print(f"You entered: {user_input}")
class AdvancedInterface(Interface):
    def get_title(self):
        return "Advanced Interface"
    def get_label_text(self):
        return "Enter a number:"
    def on_button_click(self):
        user_input = input("Enter a number: ")
        try:
            num = int(user_input)
            print(f"Square of {num} is {num**2}")
        except ValueError:
            print("Invalid input. Please enter a valid number.")
# Create an instance of SimpleInterface
simple_interface = SimpleInterface()
# Use the methods of SimpleInterface
print(simple_interface.get_title())
print(simple_interface.get_label_text())
simple_interface.on_button_click()

# Create an instance of AdvancedInterface
advanced_interface = AdvancedInterface()
# Use the methods of AdvancedInterface
print(advanced_interface.get_title())
print(advanced_interface.get_label_text())
advanced_interface.on_button_click()
```

# Strong Typing

```python
# Duck Typing
class Duck:
    def walk(self):
        print("thapak thapak thapak thapak")

class Horse:
    def walk(self):
        print("tabdak tabdak tabdak tabdak")

class Cat:
    def talk(self):
        print("Meow")

def myfunction(obj):
    if hasattr(obj, 'walk'):
        obj.walk()
    if hasattr(obj, 'talk'):
        obj.talk()

d = Duck()
myfunction(d)

h = Horse()
myfunction(h)

c = Cat()
myfunction(c)
```

```python
# Python program showing use of __call__() method
class MyDecorator:
    def __init__(self, function):
        self.function = function
    def __call__(self):
        # We can add some code before function call
        self.function()
        # We can also add some code after function call.
# adding class decorator to the function
@MyDecorator
def function():
    print("GeeksforGeeks")
function()
```

```python
class Decor(object):
    def __init__(self,func):
        self.function=func
    def __call__(self,*args):         
        try:
            if any([isinstance(i,str) for i in args]):
        #[False,False,False]
                raise TypeError("Cannot pass string as arguments")
            else:
                return self.function(*args)
        except Exception as obj:
                print(obj)
@Decor
def add(*args):
    sum1=0
    for num in args:
        sum1=sum1+num
        return sum1
print(add(10,20,30))
print(add(10,'20',30))
```

# Pass by value | Pass by reference

```python
def call_by_value(x):
    x = x * 2
    print("in function value updated to", x)
    return
num = 6
print("Initial value:", num)
call_by_value(num)
print("Final value:", num)
```

# Copy Constructor vs Assignment Operator

```python
#Copy Constructor:
class MyClass:
    def __init__(self, value):
        self.value = value

my_object = MyClass(10)
new_object = MyClass(my_object)

print(my_object.value)
print(new_object.value)
```

```python
#Assignment Operator:
class MyClass:
    def __init__(self, value):
        self.value = value

my_object = MyClass(10)
new_object = my_object

print(my_object.value) 
print(new_object.value) 
```
