```python
import requests
from bs4 import BeautifulSoup
import os

# Path to your local HTML file
file_path = r"C:\Users\acer\Downloads\Compressed\takeout-20250107T073828Z-001\Takeout\YouTube and YouTube Music\history\watch-history.html"
#'path/to/your/local_file.html'
# Open and read the local HTML file
with open(file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(content, 'html.parser')

# Extract all links (anchor tags)
links = soup.find_all('a')

# Define the phrase you are looking for
phrase = "https://www.youtube.com/watch"

# Filter and print links that contain the specific phrase
youtube_links = [link.get('href') for link in links if link.get('href') and phrase in link.get('href')]

# Print the filtered YouTube links
for yt_link in youtube_links:
    print(yt_link)
```

```python
def save_dataframe_in_chunks(df, rows_per_file, base_filename):
    total_rows = len(df)
    num_files = (total_rows + rows_per_file - 1) // rows_per_file  # Calculate the number of files

    for i in range(num_files):
        start_row = i * rows_per_file
        end_row = min((i + 1) * rows_per_file, total_rows)

        chunk = df.iloc[start_row:end_row]  # Slice the DataFrame for the current chunk

        # Generate the filename with an index
        filename = f"{base_filename}_{i + 1}.csv"

        # Save the chunk to a CSV file
        chunk.to_csv(filename, index=False)

        print(f"Saved chunk {i + 1} to {filename}")

# Example usage
# df = pd.DataFrame({"A": range(10000), "B": range(10000)})
# save_dataframe_in_chunks(df, 1000, "data_chunk")
```
