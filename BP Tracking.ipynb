```python
d=df[['BP']]
# Split 'BP' into two separate columns
d[['Upper Value', 'Lower Value']] = d['BP'].str.split('/', expand=True).astype(int)
```

```python
import matplotlib.pyplot as plt

# Plot the three numerical columns
plt.figure(figsize=(12, 6))

plt.plot(df['Date'],df['Pulse'], label='Pulse', marker='o')
plt.plot(df['Date'],d['Upper Value'], label='Upper Value', marker='o')
plt.plot(df['Date'],d['Lower Value'], label='Lower Value', marker='o')

plt.xlabel('Date')
plt.ylabel('Values')
plt.title('BP')
plt.legend()
plt.grid(True)


plt.savefig('plot.png')
plt.close()
```

```python
# Create a new Word document
doc = Document()

# Add a table to the document
table = doc.add_table(rows=df.shape[0] + 1, cols=df.shape[1])

# Add the column names to the first row of the table
for j, col_name in enumerate(df.columns):
    table.cell(0, j).text = col_name

# Add the data from the DataFrame to the table
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        table.cell(i + 1, j).text = str(df.values[i, j])
        
# Add the plot image to the DOCX document
doc.add_picture('plot.png')

# Save the document
doc.save('table2.docx')
```
