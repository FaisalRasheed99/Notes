output={1:2,2:2,3:2,4:3,7:1}

```python
l=[3,4,1,3,2,2,1,4,4,7]
```

```python
unique = []
for i in l:
    if i not in unique:
        unique.append(i)
```

```python
unique
```

```python
count = {}
for i in l:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
```
