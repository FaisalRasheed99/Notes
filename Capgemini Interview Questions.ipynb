```python
ids = [1, 2, 3]
ages = [25, 30, 22]
names = ["Alice", "Bob", "Charlie"]
```

I want to have the Following output:

{
    "Alice": {
        "id": 1,
        "age": 25
    },
    "Bob": {
        "id": 2,
        "age": 30
    },
    "Charlie": {
        "id": 3,
        "age": 22
    }

}

```python
result = {}

for i in range(len(names)):
    result[names[i]] = {"id":ids[i],"age":ages[i]}
result
```

Using map function:

```python
dict(map(lambda x:(x[2],{"id":x[0],"age":x[1]}),map(lambda i:(ids[i],ages[i],names[i]),range(len(names)))))
```
