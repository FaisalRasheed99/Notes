Check which pairs of numbers from a given list add up to 15

```python
def find_pairs_with_sum(numbers, target_sum=15):
    pairs = []
    
    for i in range(len(numbers)):
        for j in range(i + 1, len(numbers)):
            if numbers[i] + numbers[j] == target_sum:
                pairs.append((numbers[i], numbers[j]))
    
    return pairs

# Example usage
numbers = [2, 7, 13, 8, 5, 10, 1, 14]
result = find_pairs_with_sum(numbers)

print("Pairs with sum 15:", result)
```
