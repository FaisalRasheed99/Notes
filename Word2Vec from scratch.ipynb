```python
import numpy as np
from collections import defaultdict
import re

# Step 1: Data Preprocessing
def preprocess(text):
    # Tokenize text
    text = re.sub(r'[^\w\s]', '', text.lower())  # Remove punctuation and lowercase
    words = text.split()
    return words
```

```python
# Create a vocabulary and generate training data
def generate_training_data(words, window_size):
    vocab = set(words)
    word_to_index = {word: i for i, word in enumerate(vocab)}
    index_to_word = {i: word for i, word in enumerate(vocab)}
    
    training_data = []
    
    for i, word in enumerate(words):
        context_words = []
        for j in range(max(0, i - window_size), min(len(words), i + window_size + 1)):
            if i != j:
                context_words.append(word_to_index[words[j]])
        training_data.append((word_to_index[word], context_words))
    
    return training_data, word_to_index, index_to_word
```

```python
# Step 2: Initialize Parameters
def initialize_parameters(vocab_size, embedding_dim):
    W1 = np.random.uniform(-0.8, 0.8, (vocab_size, embedding_dim))
    W2 = np.random.uniform(-0.8, 0.8, (embedding_dim, vocab_size))
    return W1, W2
```

```python
# Step 3: Forward Propagation
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
def forward_pass(center_word_idx, W1, W2):
    h = W1[center_word_idx]
    u = np.dot(W2.T, h)
    y_pred = softmax(u)
    return y_pred, h, u
```

```python
# Step 4: Loss Function (Cross-Entropy)
def compute_loss(y_pred, target_idx):
    return -np.log(y_pred[target_idx] + 1e-9)  # Adding epsilon to avoid log(0)
```

```python
# Step 5: Backpropagation
def backprop(y_pred, target_idx, h, W1, W2, center_word_idx, learning_rate):
    y_true = np.zeros_like(y_pred)
    y_true[target_idx] = 1
    
    error = y_pred - y_true
    dW2 = np.outer(h, error)
    dW1 = np.dot(W2, error)
    
    W1[center_word_idx] -= learning_rate * dW1
    W2 -= learning_rate * dW2
```

```python
# Step 6: Training Loop
def train_skipgram(words, window_size, embedding_dim, learning_rate, epochs):
    training_data, word_to_index, index_to_word = generate_training_data(words, window_size)
    vocab_size = len(word_to_index)
    
    W1, W2 = initialize_parameters(vocab_size, embedding_dim)
    
    # Training
    for epoch in range(epochs):
        total_loss = 0
        
        for center_word_idx, context_word_indices in training_data:
            for target_idx in context_word_indices:
                y_pred, h, u = forward_pass(center_word_idx, W1, W2)
                loss = compute_loss(y_pred, target_idx)
                total_loss += loss
                
                backprop(y_pred, target_idx, h, W1, W2, center_word_idx, learning_rate)
        
        print(f'Epoch {epoch+1}, Loss: {total_loss:.4f}')
    
    return W1, word_to_index, index_to_word
```

```python
# Example Usage
text = "I love natural language processing and deep learning"
words = preprocess(text)
window_size = 2
embedding_dim = 10
learning_rate = 0.01
epochs = 1000

W1, word_to_index, index_to_word = train_skipgram(words, window_size, embedding_dim, learning_rate, epochs)

# Print word embeddings
for word, idx in word_to_index.items():
    print(f"Word: {word}, Embedding: {W1[idx]}")
```
